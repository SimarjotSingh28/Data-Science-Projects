```python
import numpy as np
import pandas as pd
```

```python
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('movies.csv')
```

```python
selected_features=['genres','keywords','tagline','cast','director']
```

```python
for feature in selected_features:
    df[feature]=df[feature].fillna('')
```

```python
combined_features=df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
```

```python
combined_features[0]
```

```python
vector=TfidfVectorizer()
```

```python
feature_vectors=vector.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
similarity.shape
```

```python
movie_name=input('Enter the movie name : ')
```

```python
list_of_all_titles=df['title'].tolist()
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
```

```python
print(close_match)
```

```python
index_of_movie=df[df.title==close_match]['index'].values[0]
print(index_of_movie)
```

```python
similarity_score=list(enumerate(similarity[index_of_movie]))
```

```python
len(similarity_score)
```

```python
sorted_similar_movie=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print('Movies Suggested are:\n')
i=1
for movie in sorted_similar_movie:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<30):
        print(i,".",title_from_index)
        i+=1
```

```python
movie_name=input('Enter the movie name : ')
list_of_all_titles=df['title'].tolist()
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
close_match=find_close_match[0]
index_of_movie=df[df.title==close_match]['index'].values[0]
similarity_score=list(enumerate(similarity[index_of_movie]))
sorted_similar_movie=sorted(similarity_score,key=lambda x:x[1],reverse=True)


print('Movies Suggested are:\n')
i=1
for movie in sorted_similar_movie:
    index=movie[0]
    title_from_index=df[df.index==index]['title'].values[0]
    if(i<30):
        print(i,".",title_from_index)
        i+=1
```
